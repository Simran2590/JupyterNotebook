### This script uses the random module to generate a list of 10 random integers between 1 and 100. It then uses the built-in sum() function to calculate the sum of the numbers in the list, and the len() function to calculate the number of elements in the list (which is used to calculate the average). Finally, it uses the print() function to display the list of numbers, the sum, and the average on the screen.

```python
import random

# Create a list of random numbers
numbers = [random.randint(1, 100) for x in range(10)]

# Calculate the sum of the numbers
total = sum(numbers)

# Calculate the average of the numbers
average = total / len(numbers)

# Print the results
print("Numbers:", numbers)
print("Sum:", total)
print("Average:", average)
```
